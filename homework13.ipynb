```python
import random

out_file = "110207420_practice1.csv"

ofi = open(out_file,"w",encoding="Big5")
fin = open("common_names.txt","r")

num_students = 50
names = []
count = 0

def to_GPA(n):
    if n == 0:
        return 0.0
    elif 1 <= n <= 49:
        return 0.0
    elif 50 <= n <= 59:
        return 1.0
    elif 60 <= n <= 62:
        return 1.7
    elif 63 <= n <= 66:
        return 2.0
    elif 67 <= n <= 69:
        return 2.3
    elif 70 <= n <= 72:
        return 2.7
    elif 73 <= n <= 76:
        return 3.0
    elif 77 <= n <= 79:
        return 3.3
    elif 80 <= n <= 84:
        return 3.7
    elif 85 <= n <= 89:
        return 4.0
    elif 90 <= n <= 100:
        return 4.3

for line in fin:
    s = line.strip().split()
    if len(s) > 0:
        names.append(s[1])
        count = count + 1
        if count == num_students:
            break



t1 = []
t2 = []
t3 = []
  
for i in range(50):
    sid = "1102074{:02d}".format(i+1)
    name = names[i]
    num_courses =  random.randint(3,8)
    total_credits = 0
    total_score = 0
    total_gpa_score = 0
    
    for i in range(num_courses):
        credit = random.randint(1,4)
        score = random.randint(0,100)
        GP = to_GPA(score)
        total_credits += credit
        total_score += (score * credit)
        total_gpa_score += (GP * credit)
        
    average = total_score // total_credits
    GPA = total_gpa_score // total_credits
    
    if total_credits < 10:
        continue
    else:
        t1.append((sid, name, average))
        t2.append((name, sid, average))
        t3.append((average, sid, name))
        
t1.sort()
t2.sort()
t3.sort()

ofi.write("依照學號排序\n") 
for i in t1:
    ofi.write("{:<},{},{}\n".format(i[0],i[1],i[2]))

ofi.write("依照姓名排序\n")
for i in t2:
    ofi.write("{},{},{}\n".format(i[0],i[1],i[2]))

ofi.write("依照平均成績排序\n")
for i in t3:
    ofi.write("{:<},{},{}\n".format(i[0],i[1],i[2]))
   
fin.close()
ofi.close()
print("done")
```

```python
import qrcode
from PIL import Image

in_file = "courses.csv"
fin = open(in_file)

for line in fin:
    s = line.strip()
    t = s.split(",")
    course = t[0]
    out_file = '{}.png'.format(course)
    qr_big = qrcode.QRCode(
     error_correction=qrcode.constants.ERROR_CORRECT_H,
     box_size=14,
     border=3
    )
    url_str = 'https://geomatics.nccu.edu.tw/cgi-bin/form.py?course={}'
    data = url_str.format(course)
    qr_big.add_data(data)
    qr_big.make()
    img_qr_big = qr_big.make_image().convert('RGB')
    print('QRCode image size:', img_qr_big.size)
    logo = Image.open('nccu_logo_3.jpg')
    print('NCCU logo image size:', logo.size)
    pos = ((img_qr_big.size[0] - logo.size[0]) // 2, (img_qr_big.size[1] - logo.size[1])
    // 2)
    print('pos = ', pos)
    img_qr_big.paste(logo, pos)
    img_qr_big.save(out_file)
    html = """<!DOCTYPE html>
    <html>
    <head>
    <meta http-equiv="content-type" content="text/html; charset=Big5" />
    </head>
    <body>
    <h3>科目代號(Course #): {}<br>
    科目名稱： {}<br>
    Course Name: {}<br>
    授課教師： {}<br>
    Instructor: {}<br>
    系所： {}<br>
    上課時間(Session)： {} (wed09-12)<br>
    </h3>
    <p>
    <img src="./{}.png">
    </body>
    </html>"""
    out_html = "{}.html".format(course)
    out_file = open(out_html,"w")
    out_file.write(html.format(t[0],t[2],t[3],t[4],t[5],t[6],t[8],course))
    
fin.close()
out_file.close()
```
