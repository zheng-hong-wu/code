## <font color="red">練習1</font>

```python
import random
from math import dist
f=open('names.csv')

last_names = []
first_names = []
name = []

out_file = '110207420_data.csv'
fout = open(out_file, 'w')

for i in range(100):
    aline = f.readline()
    line = aline.strip()
    delimeter=","
    s1,s2,s3 = line.split(delimeter)
    
    last_names.append(s3)
        
    first_names.append(s1)
    first_names.append(s2)
    
for i in range(150):
    l_name = random.choice(last_names)
    f_name = random.choice(first_names)
    full_name = '{},{}'.format(f_name, l_name)
    print('full_name: ', full_name)
    
    subject = random.randint(3,8)
    print("credits = ",subject)
    person = str(full_name)
    person += str(',')  
    
    for j in range(int(subject)):
        credits= random.randint(1,4)
        person += str(credits)
        person += str(",")
        
        grades = random.randint(1,100)
        person += str(grades)
        person += str(",")
    
    
    n_person = person[:-1]
    print(n_person)
    
    print(" ")  
    print("--------------------")
    print(" ") 
    
    persons = '{}'.format(n_person) 
    
    name.append(full_name)
    fout.write('{}\n'.format(persons))
     
def histogram(s):
    d = dict()
    for c in s:
        if c not in d:
            d[c] = 1
        else:
            d[c] += 1
            
    return d

print('name = ', name)
print(" ")  
print("----------CHECK----------")
print(" ") 
result = histogram(name)
print(result)

fout.close()
print(" ")  
print("--------------------")
print(" ") 
print('done')
```

## <font color="red">練習2</font>

```python
f=open('110207420_data.csv')

out_file = '110207420_grade.csv'
fout = open(out_file, 'w')

average_grades_list=[]
average_gp_list=[]

for i in range(150):
    aline = f.readline()
    line = aline.strip()
    delimeter=","
    a = line.split(delimeter)
    print(a)
    


    
    b=a[2::2]
    b_int = list(map(int, b))
    total_credits = sum(b_int)
    print("total_credits = ", total_credits)
    
    c=a[3::2]
    c_int = list(map(int, c))
    
    k=0
    total_grades=0
    for i in range(len(b)):
        grades = b_int[k]*c_int[k]
        k+=1
        total_grades += grades
    print("total_grades = ",total_grades)
    
    average_grades=total_grades/total_credits
    print("average_grades = ",average_grades)
    
    
    k=0
    c_gp=[]
    for i in range(len(b)):
        score = c_int[k]
        k+=1
        if 100>=score>=90:
            gp = 4.3
        elif 89>=score>=85:
            gp = 4.0
        elif 84>=score>=80:
            gp = 3.7
        elif 79>=score>=77:
            gp = 3.3
        elif 76>=score>=73:
            gp = 3.0
        elif 72>=score>=70:
            gp = 2.7
        elif 69>=score>=67:
            gp = 2.3
        elif 66>=score>=63:
            gp = 2.0
        elif 62>=score>=60:
            gp = 1.7
        elif 59>=score>=50:
            gp = 1.0
        else:
            gp = 0.0
            
        c_gp.append(gp)
    print("c_gp = ",c_gp)
    
    k=0
    total_gp=0
    for i in range(len(b)):
        gp = b_int[k]*c_gp[k]
        k+=1
        total_gp += gp
        
    print("totol_gp = ",round(total_gp,1))
    
    average_gp=(round(total_gp,1))/total_credits
    print("average_gp = ",average_gp)
    
    name = a[0:2]
    delimiter = ','
    name_str = delimiter.join(name)
    
    total = '{},{},{},{}'.format(name_str, total_credits,average_grades,average_gp)
    fout.write('{}\n'.format(total))
    
    print(" ")  
    print("--------------------")
    print(" ") 
    
    
    average_grades_list.append(average_grades)
    average_gp_list.append(average_gp)
    
    
fout.close()
print('done')
```
